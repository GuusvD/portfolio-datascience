```python
## Perform a bivariate analysis (Pearson correlation and scatter plot) on at least 1 combination of 2 columns
## with numeric data in the dataset that you chose in portfolio assignment 4. 
import pandas as pd
import scipy.stats as st
imdb = pd.read_csv('imdb_top_1000.csv', sep=',')
```

```python
## Pearson correlation
imdbCorr = imdb.corr()
imdbCorr.style.background_gradient(cmap='coolwarm', axis=None)
```

```python
## Scatter plot (between No_of_Votes and Meta_score)
imdb.plot(kind='scatter', x='No_of_Votes', y='Meta_score')
```

```python
## As can be seen in both the table and the scatterplot. There isn't really a strong bond between the rating and how many votes
## a movie gets. For instance a famous movie can get both high or low votes, it's all up to the viewers experience. The 
## No_of_Votes and Meta_score do however have a stronger bond that I'd expected beforehand.
```
