```python
## Do a Numerical VS Categorical bivariate analysis on the penguins dataset.
import seaborn as sns
```

```python
penguins = sns.load_dataset("penguins")
penguins.head()
```

```python
## Choose one of the categorical columns: species, island or sex
## use groupby(").mean() too look at the means of the numerical columns. Does it look like there is a difference 
## between categories?
penguins.groupby("species").mean()

## It definitely does look like there is a difference between the 3 different species. The means from "Adelie" and "Chinstrap"
## are however really similar if you were to look at the bill depth, flipper length and body mass.
```

```python
sns.barplot(y="bill_length_mm", x="species", data=penguins)

## There is a statistically significant difference in this plot between Adelie and the two other species, because the Adelie
## species bar is not overlapping the other two species (bars). The Chinstrap and Gentoo bars are however overlapping eachother
## so nothing can be said about these species.
```

```python
sns.barplot(y="bill_depth_mm", x="species", data=penguins)

## There is a statistically significant difference in this plot between Gentoo and the two other species, because the Gentoo
## species bar is not overlapping the other two species (bars). The Adelie and Chinstrap bars are however overlapping eachother
## so nothing can be said about these species.
```

```python
sns.barplot(y="flipper_length_mm", x="species", data=penguins)

## There is a statistically significant difference between all species, because they are not overlapping eachother.
```
