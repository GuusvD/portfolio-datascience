```python
import pandas as pd
imdb = pd.read_csv('imdb_top_1000.csv', sep=',')
```

```python
imdb.head()
```

```python
### Perform a univariate analysis on:
## 2 columns with categorical data
## 2 columns with numerical data

### Univariate analysis categorical (1 / 2)
## Genre
gen_plt = imdb['Genre'].head(20).value_counts().plot(kind='pie')
gen_plt.set_title('Most used genre combinations in top 20 movies')
gen_plt.set_ylabel('')
```

```python
### Univariate analysis categorical (2 / 2)
## Certificate
cer_plt = imdb['Certificate'].value_counts().plot(kind='bar')
cer_plt.set_title("Amount of certificates used per type")
cer_plt.set_ylabel("Amount")
cer_plt.set_xlabel("Type")
```

```python
### Univariate analysis on numerical data (1 / 2)
## IMDB_Rating
rat_plt = imdb['IMDB_Rating'].head(20).value_counts().plot(kind='bar')
rat_plt.set_title("Amount of IMDB ratings with same rating in the top 20 movies")
rat_plt.set_ylabel("Amount")
rat_plt.set_xlabel("Rating")
```

```python
### Univariate analysis on numerical data (2 / 2)
## Released_Year
rel_plt = imdb['Released_Year'].value_counts().plot(kind='line')
rel_plt.set_title("Amount of movies released during time period")
rel_plt.set_ylabel("Amount")
rel_plt.set_xlabel("Released year")
```
